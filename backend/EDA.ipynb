```python
Crime2023Part1 = pd.read_csv('./datasets/Crime2023Part1.csv')
Crime2023Part2 = pd.read_csv('./datasets/Crime2023Part2.csv')
Crime2024Part2 = pd.read_csv('./datasets/Crime2024Part2.csv')
Crime2024Part1 = pd.read_csv('./datasets/Crime2024Part1.csv')
Crime2018Part1 = pd.read_csv('./datasets/Crime2018Part1.csv')
```

```python
allCrimeData = pd.concat([Crime2023Part1, Crime2023Part2, Crime2024Part1, Crime2024Part2], ignore_index=True)
allCrimeData.drop(columns=['X', 'Y', 'QualityOfLife', 'ObjectId', 'Arrest', 'LarcenyCode' ], inplace=True)
```

```python
allCrimeData
```

```python
allCrimeData.isna().sum()   
```

```python
crimeWeights ={
    "LARCENY": 1,
    "BURGLARY": 2,
    "MOTOR VEHICLE THEFT": 3,
    "ROBBERY": 4,
    "AGGRAVATED ASSAULT": 5,
}

# Group the data by address and calculate the sum of crime weights
address_crime_weights = allCrimeData.groupby('ADDRESS')['CODE_DEFINED'].apply(lambda x: x.map(crimeWeights).sum())

# Assign the crime weights to the addresses
allCrimeData['CRIME_WEIGHT'] = allCrimeData['ADDRESS'].map(address_crime_weights)

allCrimeData
```

```python

import folium

# Create a map object
m = folium.Map(location=['43.067961', '-76.173723'], zoom_start=12)

for index, row in allCrimeData.iterrows():
    folium.Marker(
        location=[row['LAT'], row['LONG']],
        popup=row['CODE_DEFINED'],
        icon=folium.Icon(color='red')
    ).add_to(m)


m.save('crime_map.html')


```
