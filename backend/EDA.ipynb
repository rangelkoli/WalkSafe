```python
allCrimeData = pd.concat([Crime2023Part1, Crime2023Part2, Crime2024Part1, Crime2024Part2], ignore_index=True)
allCrimeData.drop(columns=[ 'QualityOfLife', 'ObjectId', 'Arrest', 'LarcenyCode' ], inplace=True)
```

```python
crimeWeights = {
    "LARCENY": 2,
    "MV THEFT": 2,
    "BURGLARY": 10,
    "ROBBERY": 10,
    "OFFN AGAINST FAMILY": 5,
    "DISORDERLY CONDUCT": 2,
    "SIMPLE ASSAULT": 5,
    "ALL OTHER OFFENSES": 2,
    "CRIMINAL MISCHIEF": 5,
    "DRIVING UNDER INFLUENCE": 10,
    "POSSESSION/USE DANGEROUS WEAPONS": 10,
    "FRAUD": 10,
    "POSSESSION/USE DRUG": 10,
    "KIDNAPPING": 50,
    "LOITERING": 5,
    "FORGERY/COUNTERFEIT": 10,
    "STOLEN PROPERTY": 20,
    "UNAUTHORIZED USE MV": 10,
    "COERCION": 10,
    "EXTORTION": 20,
    "PROSTITUTION PATRON/PROMOTING": 25,
    "SALE/MANUFACTURE CONTROLLED SUBSTANCE": 15,
    "POSSESSION/MARIJUANA": 15,
    "LIQUOR LAW VIOLATIONS": 20
}

# Group the data by address and calculate the sum of crime weights
address_crime_weights = allCrimeData.groupby('ADDRESS')['CODE_DEFINED'].apply(lambda x: x.map(crimeWeights).sum())

# Assign the crime weights to the addresses
allCrimeData['CRIME_WEIGHT'] = allCrimeData['ADDRESS'].map(address_crime_weights)

allCrimeData

allCrimeData.to_csv('allCrimeData.csv', index=False)
```

```python

import folium
from IPython.display import IFrame
from folium.plugins import HeatMap

# Create a map object
m = folium.Map(location=['43.067961', '-76.173723'], zoom_start=12)

smallCrimes = allCrimeData[allCrimeData['CRIME_WEIGHT'] < 100]
mediumCrimes = allCrimeData[(allCrimeData['CRIME_WEIGHT'] >= 100) & (allCrimeData['CRIME_WEIGHT'] < 500)]
largeCrimes = allCrimeData[allCrimeData['CRIME_WEIGHT'] >= 500]

# Create heatmap layers
heatmap_layer = HeatMap(data=smallCrimes[['LAT', 'LONG']], radius=10, color='green')
heatmap_layer2 = HeatMap(data=mediumCrimes[['LAT', 'LONG']], radius=10, color='orange')
heatmap_layer3 = HeatMap(data=largeCrimes[['LAT', 'LONG']], radius=10, color='red')

# Add the heatmap layers to the map
heatmap_layer.add_to(m)
heatmap_layer2.add_to(m)
heatmap_layer3.add_to(m)

# Display the map
m
```
